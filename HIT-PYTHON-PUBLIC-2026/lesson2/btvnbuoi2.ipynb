```python
import math
def demchuso(n):
    n=abs(n)
    if(n<10):
        return 1
    else:
        return 1+demchuso(n//10)
def tongchuso(n):
    n=abs(n)
    if(n<10):
        return n
    else:
        return n%10+tongchuso(n//10)
def soNT(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True 


n=int(input("Nhập số nguyên: "))
sl=demchuso(n)
tong=tongchuso(n)
lasnt=soNT(n)
print("Số lượng chữ số:",sl)    
print("Tổng các chữ số:",tong)
if lasnt:
    print(n,"là số nguyên tố")      
else:
    print(n,"không là số nguyên tố")    
```

```python
sotien=int(input("Nhập số tiền: "))

sochaibia=sotien//28
tongbia=int(sochaibia)
vochai=int(sochaibia)

while vochai>=3:
    chaimoi=vochai//3
    vodu=vochai%3

    tongbia+=chaimoi
    vochai=chaimoi+vodu

print("Tổng số chai bia có thể mua được là:",tongbia)    
```
